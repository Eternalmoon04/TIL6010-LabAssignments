```python
import pandas as pd
import datetime
import plotly.express as px
```

```python
file_path = 'flightdata.csv'
df = pd.read_csv(file_path)
```

```python
#Select airports
YSSY = df[df.origin == 'YSSY']
YSSY = YSSY.append(df[df.destination == 'YSSY'])
VHHH = df[df.origin == 'VHHH']
VHHH = VHHH.append(df[df.destination == 'VHHH'])
VIDP = df[df.origin == 'VIDP']
VIDP = VIDP.append(df[df.destination == 'VIDP'])
RJTT = df[df.origin == 'RJTT']
RJTT = RJTT.append(df[df.destination == 'RJTT'])
```

```python
RJAA = df[df.origin == 'RJAA']
RJAA = RJAA.append(df[df.destination == 'RJAA'])
RCTP = df[df.origin == 'RCTP']
RCTP = RCTP.append(df[df.destination == 'RCTP'])
VTBD = df[df.origin == 'VTBD']
VTBD = RCTP.append(df[df.destination == 'VTBD'])
```

```python
WSSS = df[df.origin == 'WSSS']
WSSS = WSSS.append(df[df.destination == 'WSSS'])
NZAA = df[df.origin == 'NZAA']
NZAA = NZAA.append(df[df.destination == 'NZAA'])
```

```python
YSSY['day'] = pd.to_datetime(YSSY['day'])
YSSY = YSSY.set_index(YSSY['day'])
YSSY['YSSY'] = 1
yssy = YSSY.resample('D').sum()
VHHH['day'] = pd.to_datetime(VHHH['day'])
VHHH = VHHH.set_index(VHHH['day'])
VHHH['VHHH'] = 1
vhhh = VHHH.resample('D').sum()
VIDP['day'] = pd.to_datetime(VIDP['day'])
VIDP = VIDP.set_index(VIDP['day'])
VIDP['VIDP'] = 1
vidp = VIDP.resample('D').sum()
RJTT['day'] = pd.to_datetime(RJTT['day'])
RJTT = RJTT.set_index(RJTT['day'])
RJTT['RJTT'] = 1
rjtt = RJTT.resample('D').sum()
RCTP['day'] = pd.to_datetime(RCTP['day'])
RCTP = RCTP.set_index(RCTP['day'])
RCTP['RCTP'] = 1
rctp = RCTP.resample('D').sum()
VTBD['day'] = pd.to_datetime(VTBD['day'])
VTBD = VTBD.set_index(VTBD['day'])
VTBD['VTBD'] = 1
vtbd = VTBD.resample('D').sum()
WSSS['day'] = pd.to_datetime(WSSS['day'])
WSSS = WSSS.set_index(WSSS['day'])
WSSS['WSSS'] = 1
wsss = WSSS.resample('D').sum()
NZAA['day'] = pd.to_datetime(NZAA['day'])
NZAA = NZAA.set_index(NZAA['day'])
NZAA['NZAA'] = 1
nzaa = NZAA.resample('D').sum()
total = pd.concat([yssy,vhhh,vidp,rjtt,rctp,vtbd,wsss,nzaa],axis=1)
fig = px.line(data_frame=total, x=total.index, y=['YSSY','VHHH','VIDP','RJTT','RCTP','VTBD','WSSS','NZAA'],markers=True)
fig = fig.update_layout(legend_title_text='Airport')
fig.show()
```
