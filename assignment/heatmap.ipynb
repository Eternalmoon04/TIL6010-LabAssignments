```python
import plotly.express as px
import pandas as pd
import folium
import webbrowser
from folium.plugins import HeatMap
```

```python
file_path = 'flightdata.csv'
df = pd.read_csv(file_path)
```

```python
dfcounts = df['origin'].value_counts()
dfcounts = pd.DataFrame(dfcounts)
airports = pd.read_csv('World_Airports.csv')
```

```python
airports = pd.DataFrame(airports, columns=['latitude_deg','longitude_deg'])
```

```python
dfmerge = pd.merge(dfcounts, airports, how='inner', left_on=dfcounts.index, right_on=airports.index)
```

```python
fig = px.density_mapbox(dfmerge, lat='latitude_deg', lon='longitude_deg', z='origin', title='Airports heatmap',radius=10,
center=dict(lat=0, lon=180), zoom=0, mapbox_style='stamen-terrain', width=1000, height=800, color_continuous_midpoint=240000)
fig.show()
```
